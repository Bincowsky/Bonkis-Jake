```python
import pandas as pd

# Load your dataset
file_path = '../data/raw/train_data.csv'  # Replace with the path to your dataset
data = pd.read_csv(file_path)

# Identify the countries with mixed missing and non-missing values for `ind_launch_date`
mixed_countries = [
"COUNTRY_89F9",
"COUNTRY_907E",
"COUNTRY_6C16",
"COUNTRY_445D",
"COUNTRY_4242",
"COUNTRY_5180",
"COUNTRY_06E4",
"COUNTRY_53A5",
"COUNTRY_4D48",
"COUNTRY_D8B0",
"COUNTRY_D5B9",
"COUNTRY_0309",
"COUNTRY_3AD0",
"COUNTRY_8DBB",
"COUNTRY_4253",
"COUNTRY_4442",
"COUNTRY_2E65",
"COUNTRY_B6AE"
]
  # Replace with the countries of interest
print(data['cluster_nl'].unique().shape)
# Analyze the proportion of missing `ind_launch_date` for each brand
# Analyze the proportion of missing `ind_launch_date` for each brand and country
missing_by_brand_country = data.groupby(['cluster_nl']).apply(
    lambda x: (x['ind_launch_date'] == '-1').sum() / len(x)
).sort_values(ascending=False).reset_index(name='perc_missing')

filtered_missing_by_brand_country= missing_by_brand_country[
    (missing_by_brand_country['perc_missing'] == 0.0) & 
    (missing_by_brand_country['perc_missing'] <= 1.0)
]

# Output the results
print("Proportion of Missing `ind_launch_date` by Brand:")
print(filtered_missing_by_brand_country)
```
